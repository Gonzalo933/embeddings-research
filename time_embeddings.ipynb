The idea is to have "color curves" over time.

Each point in each curve will be represented as a vector of three values (RGB). Curves will vary their values by sampling from a normal distribution with the mean as the current value.

So, for example, if the curve 1, at T1 has the value [123, 65, 30], it will generate three new values from three normal distributions with means 123, 65, 30 (or sampling from a standard normal dist. and adding the value). The new three values will represent the value of the curve at T2.

After generating T points, the final label will be R, G or B depending on which value has been dominant over the time period.


Then we need to sample from each curve because we cannot use all the points (too many dimensions). The question remaining is how to do that.
